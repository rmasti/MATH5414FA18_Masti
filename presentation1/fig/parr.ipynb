```python
#kernel names
fp=np.zeros(7)
dr = np.zeros(7)
dw = np.zeros(7)

at = np.zeros(7)
st = 32 # bytes per transaction
#computeSourceTerm
fp[0] = 42470400
dr[0] = 1061980
dw[0] = 736195
#computeRes
fp[1] = 19353600
dr[1] = 13096149
dw[1] = 2148463
#rungeKutta
fp[2] = 37632000
dr[2] = 3581284
dw[2] = 1529094
#muscl_lR
fp[3] = 100266638
dr[3] = 13405700
dw[3] = 4540956
#compute2dF
fp[4] = 156240000
dr[4] = 2957948
dw[4] = 1623642
#compute2dG
fp[5] = 174787200
dr[5] = 2776962
dw[5] = 1792364
#muscl_BT
fp[6] = 100239720
dr[6] = 2427089
dw[6] = 1293130

# comput arithmetic intensity

for i in np.arange(7):
    at[i] = fp[i]/(dr[i]+dw[i])*32
    print(at[i])

    


gp = fp/1.0e9; #gflops
fig, ax = plt.subplots(figsize=(10.0,5.0))
ax.loglog(at, gp, '*r')

rf = np.linspace(0.1,np.max(at)*1.5,1000)


mxgflop = rf*(549.0)+0 #549 gb/s

mxgflop = np.clip(mxgflop,0, 4700)

ax.loglog(rf, mxgflop,'b')

ax.set_xlabel("Arithmetic Intensity (FLOPS/Byte)")
ax.set_ylabel("GFLOPS/sec")


n = np.arange(7)+1

for i, txt in enumerate(n):
    ax.annotate(txt, (at[i], gp[i]))

print(gp)

plt.tight_layout()
plt.savefig("rooflinemodeling.png", dpi=300)
plt.show()
```
